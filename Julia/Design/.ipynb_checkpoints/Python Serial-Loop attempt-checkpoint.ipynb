```python
 # Import some useful libraries
import serial, time
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [18, 8]
    
for i in range(10):
   

    ser = serial.Serial('COM6:', 115200) # Usually ACM0, changes with restarts

    #%%time
    r = ser.read_all() # clear serial buffer

    # Start a conversion
    ser.write(b'c')
    while ser.in_waiting < 1:
        pass # wait for a response
    time.sleep(0.05) # This extra delay helps with reliability - it gives the micro time to send all it needs to
    r = ser.read_all()
    #print("fadgjaoigr")
    print(r) # Prints the time

    # Get the values
    ser.write(b'p') # Print DMA buffer
    while ser.in_waiting < 1:
        pass # wait for a response
    print('p')
    a = []
    i = 0
    while True:
        if ser.in_waiting < 10:
            print(ser.in_waiting)
        if ser.in_waiting < 5:
            time.sleep(0.005) # Wait and check again
            #print()
            if ser.in_waiting < 5:
                break
        b = ser.readline()
        a.append(b)
        i += 1
    
    print(len(a), i) # Should be 10000 readings

    ser.close() # Close the connection when you're done.

    kcount=0;
    for k in a:
        g = k.decode('ascii')
        if (g.count('\r') == 2):
            a.pop(kcount)
        #kkk[k]=kkk[k-1];
        #print(g);
        if (int(k.decode('utf-8'))>5000):
            a.pop(kcount)
        
        kcount += 1
        

    #print(a[0:10])
    print(len(a))
    #round(((int(n)-2047.5)/2047.5),6)
    f = [round(((int(n)-2047.5)/2047.5),6) for n in a]
    #print(f[0:10])
    
    plt.plot(f)
    h= open("Waveform.txt","w+")

    #test =  {'2', '1', '3'}
    s = ', '
    tempCount = 0;
    for i in f:
        #print(i)
        if(tempCount< len(f)-1):
            h.write(str(i)+", ")
        else:
            h.write(str(i))
        tempCount +=1
    h.close() 
    #print(f)

    time.sleep(3)

#for j in f:
#    print(a)
#    
```
